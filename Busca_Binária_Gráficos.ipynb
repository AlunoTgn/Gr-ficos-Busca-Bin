```python
import matplotlib.pyplot as plt

# Dados de exemplo
tamanhos_sequencia = [10**4, 10**5, 10**6, 10**7]
resultados_binaria = [[0.1, 0.5, 1.2, 5.3],   # Resultados para 100 consultas
                      [0.2, 0.7, 1.8, 7.5],   # Resultados para 1000 consultas
                      [0.5, 1.3, 3.5, 12.2],  # Resultados para 10000 consultas
                      [1.0, 2.5, 7.0, 25.0]]  # Resultados para 100000 consultas

# Criando os gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plotando os resultados para diferentes números de consultas
for i in range(2):
    for j in range(2):
        index = 2*i + j
        axs[i, j].plot(tamanhos_sequencia, resultados_binaria[index])
        axs[i, j].set_title(f'{10**(index + 2)} Consultas')
        axs[i, j].set_xlabel('Tamanho da Sequência')
        axs[i, j].set_ylabel('Tempo Médio de Execução (segundos)')
        axs[i, j].set_xscale('log')
        axs[i, j].grid(True)

plt.tight_layout()
plt.show()




```
